# (B) Comparative Analysis of Datasets

**Supported data mining tasks**
- ImageNet-1k: Large-scale ML, Clustering, and Visual Data Mining
- WikiArt: Large-scale ML, Clustering, and Visual Data Mining
- AIDA: Large-scale ML, Clustering, and Visual Data Mining

**Data Quality Issues**
- ImageNet-1k: Label noise, overly specific classes, near-duplicate images
- WikiArt: Style and genre labels can overlap considerably, large class imbalance, image quality inconsistencies, metadata inconcistencies (ex: artist names not normalized)
- AIDA: Class imbalance, overlapping classes, small amount of data

**Algorithmic feasibility**\
Being in the vision modality, computation is going to be a concern with all of these datasets. However, the AIDA dataset is clearly the smallest, and the ImageNet-1k dataset is clearly the biggest. Any one of these datasets can be filtered down to only a few select classes which will help with computation

**Bias considerations**
- ImageNet-1k: Cultural bias (western-centric objects/scenes)
- WikiArt: Cultural bias (strong Eruopean modernist focus)
- AIDA: Moderism/professional bias (most images are of modern and professional buildings)

**Ethical considerations**\
Pursuing this analysis won't harm anyone. The main thing we ought to be careful of is ensuring the art displayed in this analysis is non-offensive and is appropriate for a professional setting.

```python
import matplotlib.pyplot as plt
from datasets import load_dataset
import pandas as pd
```

```python
# Load the HugGAN WikiArt dataset
wikiart = load_dataset("huggan/wikiart")
```

## EDA

```python
# General Dataset Info
print('General Dataset Info:')
print(wikiart)

# Label EDA
df = wikiart['train'].remove_columns(["image"]).to_pandas()

print('\n')
print('=================================================================')
print('=================================================================')
print('\n')
print('First 5 labels')
display(df.head())

# Most Common labels
print('\n')
print('=================================================================')
print('=================================================================')
print('\n')
print('Most Common Labels:')
print('Artist')
display(df["artist"].value_counts().reset_index(drop=False).sort_values(by='count', ascending=False).head(20))
print('Genre')
display(df["genre"].value_counts().reset_index(drop=False).sort_values(by='count', ascending=False).head(20))
print('Style')
display(df["style"].value_counts().reset_index(drop=False).sort_values(by='count', ascending=False).head(20))


print('\n')
print('=================================================================')
print('=================================================================')
print('\n')
print('Artist distribution')
# Artist distribution
df["artist"].value_counts().head(10).plot(kind="bar")
plt.title("Most Common Artists")
plt.xlabel("Artist")
plt.ylabel("Count of Art Pieces")
plt.show()

print('\n')
print('=================================================================')
print('=================================================================')
print('\n')
print('Genre distribution')
# Style distribution
df["genre"].value_counts().head(10).plot(kind="bar")
plt.title("Most Common Genres")
plt.xlabel("Genre")
plt.ylabel("Count of Art Pieces")
plt.show()

print('\n')
print('=================================================================')
print('=================================================================')
print('\n')
print('Style distribution')
# Style distribution
df["style"].value_counts().head(10).plot(kind="bar")
plt.title("Most Common Styles")
plt.xlabel("Style")
plt.ylabel("Count of Art Pieces")
plt.show()


print('\n')
print('=================================================================')
print('=================================================================')
print('\n')
print('Image Examples')
# Image Examples
for i in range(5):
    plt.imshow(wikiart['train'][i]["image"])
    plt.title(f"{wikiart['train'][i]['artist']} | {wikiart['train'][i]['style']} | {wikiart['train'][i]['genre']}")
    plt.suptitle("Artist | Style | Genre")
    plt.axis("off")
    plt.show()

# Image size analysis
sizes = [wikiart['train'][i]["image"].size for i in range(1000)]
widths = [s[0] for s in sizes]
heights = [s[1] for s in sizes]
print('\n')
print('=================================================================')
print('=================================================================')
print('\n')
print(f'Widths of images')
display(pd.Series(widths).describe())
print(f'Heights of images')
display(pd.Series(heights).describe())

#Number of unlabeled images
print('\n')
print('=================================================================')
print('=================================================================')
print('\n')
print('Number of unlabeled images:')
display(df.isna().sum())
```
